```python
import pandas as pd
import numpy as np
from functools import reduce

print("Task 1: Create DataFrame")

df = pd.DataFrame({
        'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 35, 28],
        'City': ['New York', 'San Francisco', 'Los Angeles', 'Chicago']})

print(df)
print(df.dtypes)  # Check data types



print("\n===== Task 2: Drop Column =====")

df_dropped = df.drop(columns=['City'])
print(df_dropped)



print("\n===== Task 3: Handling Null Values =====")

df_null = pd.DataFrame({
        'A': [1, np.nan, 3],
        'B': [np.nan, 5, 6],
        'C': [7, 8, np.nan]})

print("Original DataFrame:\n", df_null)

# Fill null values
df_filled = df_null.fillna(0)
print("\nFilled with 0:\n", df_filled)

# Alternative: fill with column mean
df_mean = df_null.fillna(df_null.mean())
print("\nFilled with mean:\n", df_mean)


print("\n===== Task 4: GroupBy and Describe =====")

df_group = pd.DataFrame({
        'Category': ['A', 'B', 'A', 'B', 'A', 'C'],
        'Value': [10, 20, 15, 25, 30, 35]})

grouped = df_group.groupby('Category')['Value'].describe()
print(grouped)

print("\nInterpretation:")
print("Each category shows count, mean, std, min, max etc.")



print("\n===== Task 5: Concatenation and Merging =====")

df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})
df3 = pd.DataFrame({'C': [9, 10], 'D': [11, 12]})

# Vertical concatenation
concat_df = pd.concat([df1, df2], ignore_index=True)
print("Concatenated:\n", concat_df)

# Horizontal merge (by index)
merged_df = pd.concat([concat_df, df3], axis=1)
print("\nMerged horizontally:\n", merged_df)
print("Shape:", merged_df.shape)



print("\n===== Task 6: Tuples and Sets =====")

fruits = ('apple', 'banana', 'cherry')
numbers_set = {1, 2, 3, 4, 5}

print("Tuple:", fruits)
print("Set:", numbers_set)


fruits += ('orange',)  # creates new tuple
print("Updated tuple:", fruits)
numbers_set.add(6)
print("Updated set:", numbers_set)
print("\nExplanation: Tuples are immutable, sets are mutable.")



print("\n===== Task 7: Dictionaries =====")

students = {'John': 80, 'Jane': 90, 'Bob': 75}
print("Original:", students)

# Update score
students['John'] = 85

# Add new student
students['Alice'] = 95

print("Updated:", students)



print("\n===== Task 8: Functions and Lambda =====")

def square(x):
  return x ** 2

square_lambda = lambda x: x ** 2

print("Function:", square(4), square(6))
print("Lambda:", square_lambda(4), square_lambda(6))



print("\n===== Task 9: Iterators and Generators =====")
try:
    # Iterator class
    class EvenIterator:
        def __init__(self, limit):
            self.limit = limit
            self.current = 0
            self.count = 0

        def __iter__(self):
            return self

        def __next__(self):
            if self.count >= self.limit:
                raise StopIteration
            result = self.current
            self.current += 2
            self.count += 1
            return result

    iterator = EvenIterator(5)
    print("Iterator:", list(iterator))

    # Generator
    def even_generator(n):
        for i in range(n):
            yield i * 2

    print("Generator:", list(even_generator(5)))
except Exception as e:
    print("Error with iterator/generator:", e)


print("\n===== Task 10: Map, Reduce, Filter =====")
try:
    numbers = [1, 2, 3, 4, 5]

    squared = list(map(lambda x: x**2, numbers))
    product = reduce(lambda x, y: x * y, numbers)
    evens = list(filter(lambda x: x % 2 == 0, numbers))

    print("Squared:", squared)
    print("Product:", product)
    print("Evens:", evens)
except Exception as e:
    print("Error in map/reduce/filter:", e)


print("\n===== Task 11: OOP Rectangle Class =====")
try:
    class Rectangle:
        def __init__(self, length, width):
            if length <= 0 or width <= 0:
                raise ValueError("Length and width must be positive")
            self.length = length
            self.width = width

        def area(self):
            return self.length * self.width

        def perimeter(self):
            return 2 * (self.length + self.width)

    r1 = Rectangle(5, 3)
    r2 = Rectangle(7, 4)

    print("Rectangle 1 Area:", r1.area())
    print("Rectangle 1 Perimeter:", r1.perimeter())
    print("Rectangle 2 Area:", r2.area())
    print("Rectangle 2 Perimeter:", r2.perimeter())
except Exception as e:
    print("Error in OOP:", e)


print("\n===== Task 12: Pandas Data Analysis =====")
try:
    df_employees = pd.DataFrame({
        'Name': ['John', 'Jane', 'Bob', 'Alice', 'Charlie'],
        'Department': ['IT', 'HR', 'IT', 'Finance', 'HR'],
        'Salary': [55000, 65000, 70000, 60000, 58000]
    })

    # Average salary by department
    avg_salary = df_employees.groupby('Department')['Salary'].mean()
    print("Average Salary:\n", avg_salary)

    # Employees with salary > 60000
    high_salary = df_employees[df_employees['Salary'] > 60000]['Name']
    print("\nEmployees with salary > 60000:\n", high_salary.tolist())

    # Add Bonus column
    df_employees['Bonus'] = df_employees['Salary'] * 0.10
    print("\nUpdated DataFrame:\n", df_employees)

except Exception as e:
    print("Error in data analysis:", e)
```
